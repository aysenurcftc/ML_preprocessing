```python
data = pd.read_csv('student_performance.csv')
```

```python
x=data.drop(['G3'], axis=1)
y=data[['G3']]

sayisal_degiskenler = data.select_dtypes(include=[np.int64]).columns
kategorik_degiskenler = data.select_dtypes(exclude=[np.int64]).columns
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
#değişkenleri sayısal ve kategorik olarak gruplayabiliriz. Sayısal değişkenleri de kesikli ve sürekli olarak ayrılır.
#sürekli değişkenlere örnek -> zaman, sıcaklık
#kesikli örnek -> kütüphanedeki farklı türledeki kitap sayıları
#Kategorik değişkenleri ise nominal ve ordinal olarak ayırabiliriz.
#Ordinalde sıralamanın bir önemi vardır. Örnek sınav notlandırma(A, B, C gibi)
#Nominalde sıralamanın bir önemi yoktur.
# for one hot encoding with sklearn
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
#one hot encoding ile her kategori o kategoriye ait olduğunu gösteren bir sütün olacak şekilde
# binary(0, 1) vektörlere dönüştürülür.

enc = OneHotEncoder(categories='auto',
                        drop='first',
                        sparse_output=False)
enc.fit(X_train[kategorik_degiskenler])
X_train_enc=enc.transform(X_train[kategorik_degiskenler])
X_test_enc=enc.transform(X_test[kategorik_degiskenler])
X_train_enc.head()
```

```python
#veriler ordinal olsaydı
encoder=OrdinalEncoder()

encoder.fit(X_train[kategorik_degiskenler])
X_train_enc=encoder.transform(X_train[kategorik_degiskenler])
X_test_enc=encoder.transform(X_test[kategorik_degiskenler])
X_train_enc.head()
```
