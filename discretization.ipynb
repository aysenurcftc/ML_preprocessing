```python
#Discretization(Ayrıklaştırma) -> Sürekli verileri bir dizi bitişik aralık veya bölme oluşturarak kesikli değişkenlere
#dönüştürme işlemidir.

from sklearn import set_config
set_config(transform_output="pandas")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
```

```python
from ucimlrepo import fetch_ucirepo
rice_cammeo_and_osmancik = fetch_ucirepo(id=545)
# veri (pandaların veri çerçeveleri olarak)
pd.concat([rice_cammeo_and_osmancik['data']['features'], rice_cammeo_and_osmancik['data']['targets']], axis=1).to_csv('rice_cammeo_and_osmancik.csv', index=False)
```

```python
data = pd.read_csv('rice_cammeo_and_osmancik.csv')
```

```python
data.tail(3)
```

```python
data['Class'].replace(to_replace={'Cammeo':1, 'Osmancik':0}, inplace=True)
```

```python
data.tail(3)
```

```python
data.describe()
```

```python
x=data.drop('Class', axis=1)
y=data[['Class']]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.KBinsDiscretizer.html
from sklearn.preprocessing import KBinsDiscretizer

#'uniform': Her bir özellikteki tüm bölümler aynı genişliğe sahiptir.
discretizer= KBinsDiscretizer(n_bins=10, encode='ordinal',strategy='uniform')


discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
discretizer.transform(X_test[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])


tmp = discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
tmp['Major_Axis_Length'].value_counts()
```

```python
#'quantile': Her bir özellikteki tüm bölümler aynı sayıda noktaya sahiptir.
from sklearn.preprocessing import KBinsDiscretizer

discretizer= KBinsDiscretizer(n_bins=10, encode='ordinal',strategy='quantile')

discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
discretizer.transform(X_test[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
```

```python
tmp = discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
tmp['Perimeter'].value_counts()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
#aralıklar k-ortalamalar algoritması tarafından tanımlanan kümelerdir
discretizer= KBinsDiscretizer(n_bins=10, encode='ordinal',strategy='kmeans')

discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
discretizer.transform(X_test[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
```

```python
tmp = discretizer.fit_transform(X_train[['Perimeter','Major_Axis_Length','Minor_Axis_Length', 'Eccentricity', 'Extent']])
tmp['Perimeter'].value_counts()
```

```python
#Ayrıklaştırma işlemlerinde, özelliklerin değerleri belirli aralıklara bölünür ve
#bu aralıklar karar ağacı tarafından kullanılarak tahminler yapılır.
#Bu, sürekli özellikleri ayrıklaştırmanın bir yoludur ve karar ağacının sürekli verileri işlemesine olanak tanır.

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

from sklearn.tree import DecisionTreeRegressor

discretizer= DecisionTreeRegressor(max_depth=5, random_state=42)

discretizer.fit(X_train[['Perimeter']], y_train)
X_train['Perimeter_tree'] = discretizer.predict(X_train[['Perimeter']])
X_test['Perimeter_tree'] = discretizer.predict(X_test[['Perimeter']])
```
